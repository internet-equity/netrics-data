```python
def split_hour(x):
    row = x
    return row[11:13]

Casas_device_avg.loc[:,"hour"] = Casas_device_avg.loc[:,"2021-10-27 02:51:03"].apply(split_hour)
```

```python
Casas_device_avg
```

```python
Casas_device_avg.hour = Casas_device_avg['hour'].astype('int')
print(Casas_device_avg.dtypes)
```

```python
Casas_device_avgx = Casas_device_avg['hour'].min()
Casas_device_avgx
```

```python
Casas_device_avgy = Casas_device_avg['hour'].max()
Casas_device_avgy
```

```python
fig,ax =  plt.subplots(figsize = ( 10, 8))
sns.lineplot( x = "hour", y = "37.803",
             color = 'red', data = Casas_device_avg,
             ax = ax)
plt.xticks(np.arange(min(Casas_device_avgx), max(Casas_device_avgy)+1, 1.0))
```

```python
Casas_device_avgm = Casas_device_avg.groupby('hour')[['37.803']].max()
Casas_device_avgm
```
