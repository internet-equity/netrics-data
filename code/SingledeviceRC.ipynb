```python
import pandas as pd
import numpy as np
rc_device = pd.read_csv('/srv/data/my_shared_data_folder/internet-equity/device_data_oct21/nm-mngd-20210518-075ab2f0.csv')
rc_device
```

```python
atlanta_device = rc_device.loc[rc_device['Atlanta'] == 'Atlanta']
atlanta_device
```

```python
atlanta_device_max = atlanta_device.loc[atlanta_device['avg'] == 'max']
atlanta_device_max
```

```python
atlanta_device_min = atlanta_device.loc[atlanta_device['avg'] == 'min']
atlanta_device_min
```

```python
atlanta_device_mdev = atlanta_device.loc[atlanta_device['avg'] == 'mdev']
atlanta_device_mdev
```

```python
atlanta_device_avg = atlanta_device.loc[atlanta_device['avg'] == 'avg']
atlanta_device_avg
```

```python
import pandas as pd
import matplotlib.pyplot as plt
#Making separate plots for the four datasetts
plt.figure(figsize=(10, 6))
#Dataset 1(max)
plt.subplot(2, 2, 1) 
plt.plot(atlanta_device_max['2021-10-16 03:40:58'], atlanta_device_max['37.488'])
plt.xlabel('Time')
plt.ylabel('Data')
plt.title('Atl Max Lat.')
#Dataset 2(min)
plt.subplot(2, 2, 2)
plt.plot(atlanta_device_min['2021-10-16 03:40:58'], atlanta_device_min['37.488'])
plt.xlabel('Time')
plt.ylabel('Data')
plt.title('Atl Min Lat.')
#Dataset 3(mdev)
plt.subplot(2, 2, 3)
plt.plot(atlanta_device_mdev['2021-10-16 03:40:58'], atlanta_device_mdev['37.488'])
plt.xlabel('Time')
plt.ylabel('Data')
plt.title('Atl Mdev Lat.')
#Dataset 4(avg)
plt.subplot(2, 2, 4)
plt.plot(atlanta_device_avg['2021-10-16 03:40:58'], atlanta_device_avg['37.488'])
plt.xlabel('Time')
plt.ylabel('Data')
plt.title('Atl Avg Lat.')
#fix spacing between subplots
plt.tight_layout()
#Show plots
plt.show()
#rotate the ticks
plt.xticks(rotation=45)
```
