```python
#we wrote a function to seperate the date and time then apply a date column and a time column..
#i dont know where the code went that I seperated and added the data column. but its the same function I did for the time split.
def split_hour(x):
    row = x
    return row[11:13]

atlanta_device_avg.loc[:,"hour"] = atlanta_device_avg.loc[:,"2021-10-16 03:40:58"].apply(split_hour)

```

```python
atlanta_device_avg.hour = atlanta_device_avg['hour'].astype('int')
print(atlanta_device_avg.dtypes)
```

```python
#atlanta_device_avgx means the min hour 
atlanta_device_avgx= atlanta_device_avg['hour'].min()
atlanta_device_avgx
```

```python
#atlanta_device_avgy means the max hour 
atlanta_device_avgy = atlanta_device_avg['hour'].max()
atlanta_device_avgy
```

```python
#atlanta_device_avgm means grouping by the latency of a specfic hour and finding the max
atlanta_device_avgm = atlanta_device_avg.groupby('hour')[['37.488']].max()
atlanta_device_avgm
```
