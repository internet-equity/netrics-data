```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import seaborn.objects as so
```

```python
columns = ['time','deviceid','tool','direction','protocol','target','pktloss','method','zip','isp','value','topic','annonipaddr','ipaddrchanged']
```

```python
matt_device = pd.read_csv('/srv/data/my_shared_data_folder/internet-equity/device_data_oct21/nm-mngd-20210518-005bb669.csv', names=columns)
matt_device
```

```python
atlanta_server = matt_device.loc[matt_device['target'] == 'Atlanta']
atlanta_server#atlanta_server.ping_latency.unique()
```

```python
atlanta_max = atlanta_server.loc[atlanta_server['method'] == 'max']
atlanta_max
```

```python
def split_hour(x):
    row = x
    return row[11:13]

def split_day(x):
    row = x
    return row[5:10]

atlanta_max["hour"] = atlanta_max["time"].apply(split_hour)
atlanta_max["date"] = atlanta_max["time"].apply(split_day)

atlanta_max
```
