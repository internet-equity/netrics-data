```python
#created a variable named all_files containing all 12 files
all_files = ["filtered_netrics_latency_202201.csv", "filtered_netrics_latency_202207.csv",
             "filtered_netrics_latency_202202.csv", "filtered_netrics_latency_202208.csv",
            "filtered_netrics_latency_202203.csv",  "filtered_netrics_latency_202209.csv",
            "filtered_netrics_latency_202204.csv",  "filtered_netrics_latency_202210.csv",
            "filtered_netrics_latency_202205.csv",  "filtered_netrics_latency_202211.csv",
            "filtered_netrics_latency_202206.csv",  "filtered_netrics_latency_202212.csv"]
```

```python
#created a variable name containing path for the 12 files
path="/srv/data/my_shared_data_folder//iet_2022/"
```

```python
#created a variable name containing the correct column names
columns = ['time','deviceid','tool','direction','protocol','target','pktloss','method','zip','isp','value','topic','annonipaddr','ipaddrchanged']
```

```python
#read in all 12 files under dataframe "device_2022"
device_2022 = pd.read_csv(path + 'filtered_netrics_latency_202201.csv',names=columns)
device_2022
```

```python
#for loop in order to read in each file
for file in all_files:
    df = pd.read_csv(path + file, names=columns)
    device_2022 = pd.concat([device_2022,df])
```

```python
#sorted vales
device_2022 = device_2022.sort_values(by=['time'])
device_2022.tail()
```

```python
#Located all the average latency values
device_22avg = device_2022.loc[device_2022['method'] == 'avg']
device_22avg
```

```python
#Located only 3 servers: Atlanta, Seattle, and Chicago
device_2022 = device_22avg.loc[device_22avg['target'].isin(['Atlanta', 'Seattle', 'Chicago'])]
device_2022.head()
```

```python
#Function to split hour from the time column
def split_hour(time):
    row = time
    return row[11:13]

device_2022.loc[:,"hour"] = device_2022.loc[:,"time"].apply(split_hour)
```

```python
#Function to split date from time column
def split_date(time):
    row = time
    return row[:10]

device_2022.loc[:,"date"] = device_2022.loc[:,"time"].apply(split_date) 
```

```python
#Function to split month from time column
def split_month(x):
    row = x
    return row[5:7]

device_2022.loc[:,"month"] = device_2022.loc[:,"time"].apply(split_month) 
```

```python
#Called dataframe
device_2022
```

```python
#Chagned the hour column from a string to integer
device_2022.hour = device_2022['hour'].astype('int')
print(device_2022.dtypes)
```

```python
#Created an anomaly detection function using the mean and one standard deviation
def anomaly_detec(x):
        stdev_threshold = x.std() + x.mean()
        print(stdev_threshold)
        anomaly_total = x >= stdev_threshold
        return anomaly_total
    
    
anomaly_detec(device_2022['value'])

#Created a new column "anomalytotal" which applies the anomaly detection function to the value column
device_2022["anomalytotal"]=anomaly_detec(device_2022['value'])

#device_5["anomalytotal"] = device_5["value"].apply(anomaly_detec)
```

```python
#Grouped month, date, and target from the device 2022 dataframe. 
anomaly_22 = device_2022.groupby(['month','date','target'])[['anomalytotal']].sum()
anomaly_22.head()
```

```python
#Grouped by just month and target, then pulled the monthly median anomaly totals
anomalymonth_22 = anomaly_22.groupby(['month','target'])[['anomalytotal']].median()
anomalymonth_22
```
