```python
import pandas as pd
import numpy as np
```

```python
device_2022 = pd.read_csv('/srv/data/my_shared_data_folder//iet_2022/filtered_netrics_latency_202205.csv', header=None)
device_2022
```

```python
device_2022.rename(columns={0:'time', 1:'deviceid', 2:'tool', 3:'direction', 4:'protocol', 5:'target', 6:'pktloss', 7:'method', 8:'zip', 9:'isp', 10:'value', 11:'topic', 12:'annon:paddar', 13:'ipaddnchanged'}, inplace = True)
device_2022
```

```python
device_2022 = device_2022.loc[device_2022['target'].isin(['Atlanta','Chicago','Seattle'])]
device_2022
```

```python
#Created a function which split hour from time column.
def split_hour(time):
    row = time
    return row[11:13]

device_2022.loc[:,"hour"] = device_2022.loc[:,"time"].apply(split_hour)
```

```python
#Created a function which split date from time column. 
def split_date(time):
    row = time
    return row[:10]

device_2022.loc[:,"date"] = device_2022.loc[:,"time"].apply(split_date) 
```

```python
device_2022
```

```python
#Converted hour column from string to integer to be able to plot the values. 
device_2022.hour = device_2022['hour'].astype('int')
print(device_2022.dtypes)
```

```python
device_22avg = device_2022.loc[device_2022['method'] == 'avg']
device_22avg
```

```python
def anomaly_detec(x):
        stdev_threshold = x.std() + x.mean()
        print(stdev_threshold)
        anomaly_total = x >= stdev_threshold
        return anomaly_total
    
    
anomaly_detec(device_2022['value'])

device_2022["anomalytotal"]=anomaly_detec(device_2022['value'])

#device_5["anomalytotal"] = device_5["value"].apply(anomaly_detec)
```

```python
anomaly_df = device_2022.groupby('date')[['anomalytotal']].sum()
anomaly_df
```
