```python
import pandas as pd
import numpy as np
```

```python
#imported data and renamed headers which were imported as a row before.
device_5 = pd.read_csv('/srv/data/my_shared_data_folder/internet-equity/device_data_oct21/nm-mngd-20210518-221fcce5.csv', header=None)
device_5
```

```python
device_5.rename(columns={0:'time', 1:'deviceid', 2:'tool', 3:'direction', 4:'protocol', 5:'target', 6:'pktloss', 7:'method', 8:'zip', 9:'isp', 10:'value', 11:'topic', 12:'annon:paddar', 13:'ipaddnchanged'}, inplace = True)
device_5
```

```python
device_5atl = device_5.loc[device_5['target'] == 'Atlanta']
device_5atl
```

```python
#Created a function which split hour from time column.
def split_hour(time):
    row = time
    return row[11:13]

device_5atl.loc[:,"hour"] = device_5atl.loc[:,"time"].apply(split_hour) 
```

```python
device_5atl
```

```python
#Created a function which split date from time column. 
def split_date(time):
    row = time
    return row[:10]

device_5atl.loc[:,"date"] = device_5atl.loc[:,"time"].apply(split_date) 
```
