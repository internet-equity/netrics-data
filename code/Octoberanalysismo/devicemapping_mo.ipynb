```python
atlanta_avg = matt_device.loc[matt_device['avg'] == 'avg']
atlanta_avg
```

```python
def split_hour(x):
    row = x
    return row[11:13]

atlanta_avg.loc[:,"hour"] = atlanta_avg.loc[:,"2021-10-16 15:21:00"].apply(split_hour)
```

```python
atlanta_avg
```

```python
atlanta_avg.hour = atlanta_avg['hour'].astype('int')
```

```python
atlanta_avgx = atlanta_avg['hour'].min()
atlanta_avgx
```

```python
atlanta_avgy = atlanta_avg['hour'].max()
atlanta_avgy
```

```python
atlanta_avgm = atlanta_avg.groupby('hour')[['24.955']].max()
atlanta_avgm
```

```python
fig,ax =  plt.subplots(figsize = ( 10, 8))
sns.lineplot( x = "hour", y = "24.955",
             color = 'red', data = atlanta_avgm,
             ax = ax)
plt.xticks(np.arange(min(atlanta_avgx), max(atlanta_avgy)+1, 1.0))
```
